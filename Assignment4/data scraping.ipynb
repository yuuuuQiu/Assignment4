```python
from bs4 import BeautifulSoup
import requests  
import dateutil
import csv
import pandas as pd

root_url='https://www.showstart.com/event/list?pageNo=1&pageSize=20&cityCode=21&sortType=9&showStyle='
showstyles = [1,2,3,4,6,9,23,25,26]
header= ['music #','name','performer','price','venue']
data=[] 

for i, style in enumerate(showstyles):
    url = root_url + str(style)
    print("fetching data from: {}".format(url))
    request = requests.get(url)
    soup= BeautifulSoup(request.text, 'html.parser')
    for a in soup.find_all("a", href=True, attrs={"class":{"show-item item"}}):
        music_no=str(style)
        name=a.find('div',attrs={'class':'title'})
        performer=a.find('div',attrs={'class':'artist'})
        price=a.find('div',attrs={'class':'price'})
        venue=a.find('div',attrs={'class':'addr'})
        data.append([music_no, name.text, performer.text, price.text, venue.text])

df = pd.DataFrame(data, columns = header)
df.to_csv('show_info.csv', sep='\t', encoding='utf-8')
df
```
